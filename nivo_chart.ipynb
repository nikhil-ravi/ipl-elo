```python
import pandas as pd
```

```python
rank_df = pd.read_csv("./results/rank.csv", index_col=[0])
rank_df.columns.name = "Date"
rank_df = rank_df.T
# rank.index = pd.to_datetime(rank.index)
# rank.index = rank.index.strftime("%Y-%m-%d")
```

```python
def to_nivo(df):
    df_json = df.to_dict()
    nivo = []
    for team, ranks in df_json.items():
        team_dict = {"id": team, "data": []}
        for date, rank in ranks.items():
            team_dict["data"].append({"x": date, "y": rank})
        nivo.append(team_dict)
    return nivo
```

```python
with open("./nivo_data/all_days_rank.json", "w") as f:
    f.write(str(to_nivo(rank_df)).replace("'", '"').replace("nan", "null"))
```

```python
season_end_rank = rank_df.copy()
season_end_rank.index = pd.to_datetime(season_end_rank.index)
season_end_rank = season_end_rank.resample("Y").last()
season_end_rank.index = season_end_rank.index.strftime("%Y")
```
